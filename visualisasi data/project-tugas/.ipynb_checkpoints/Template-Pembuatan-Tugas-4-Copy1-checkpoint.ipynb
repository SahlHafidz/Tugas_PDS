# Soal No 3

Berikut ini adalah data tugas mahasiswa untuk setiap pertemuan yang lulus.

1. Pertemuan 1 : 65 Mahasiswa
2. Pertemuan 2 : 30 Mahasiswa
3. Pertemuan 3 : 43 Mahasiswa
4. Pertemuan 4 : 49 Mahasiswa
5. Pertemuan 5 : 20 Mahasiswa

Buatlah layar keluaran menggunakan visualisasi PIE Chart, dengan ketentuan : 
1. Tampilkan Nama Pertemuan
2. Tampilkan Prosentase

```python
import matplotlib.pyplot as plt

# Data pertemuan dan jumlah mahasiswa yang lulus
pertemuan = ['Pertemuan 1', 'Pertemuan 2', 'Pertemuan 3', 'Pertemuan 4', 'Pertemuan 5']
jumlah_lulus = [65, 30, 43, 49, 20]
colors= ['red', 'Yellow', 'green', 'blue', 'orange']

# Membuat PIE Chart
plt.figure(figsize=(10,10))
plt.pie(jumlah_lulus, labels=pertemuan, autopct='%1.1f%%', startangle=90, colors=colors)

# Menambahkan judul
plt.title('Persentase Jumlah Mahasiswa Lulus/Pertemuan')

# Menampilkan PIE Chart
plt.show()
```

# Soal No 4

Buatlah Layar Keluaran visualisasi PIE CHART Berdasarkan Data File Excel "datamahasiswalulus"

Ketentuan : 
1. Pertemuan 1 Berwarna Merah
2. Pertemuan 2 Berwarna Hijau
3. Pertemuan 3 Berwarna kuning
4. Pertemuan 4 Berwarna abu
5. Pertemuan 5 Berwarna biru
6. Pertemuan 6 Berwarna Merah Muda

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari file Excel
file_path = '../project-tugas/datamahasiswalulus.xlsx'  # Ganti path sesuai lokasi file
df = pd.read_excel(file_path)

# Hitung jumlah mahasiswa lulus untuk setiap pertemuan
jumlah_lulus = df.groupby('Nama Pertemuan')['Jumlah Mahasiswa Lulus'].sum()

# Warna untuk setiap pertemuan sesuai dengan ketentuan
colors = ['red', 'green', 'yellow', 'gray', 'blue', 'lightcoral']

# Nama pertemuan
labels = jumlah_lulus.index

# Plot PIE CHART
plt.figure(figsize=(10, 10))
plt.pie(jumlah_lulus, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Persentase Mahasiswa Lulus Berdasarkan Pertemuan')
plt.show()
```

# Soal No 5

Buatlah Layar Keluaran visualisasi Diagram Batang Berdasarkan Data File Excel "datamahasiswalulus"

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari file Excel
file_path = '../project-tugas/datamahasiswalulus.xlsx'  # Gantilah dengan path sesuai lokasi file Anda
df = pd.read_excel(file_path)

# Tentukan data yang akan diplot (misalnya, jumlah mahasiswa lulus per pertemuan)
data_to_plot = df.groupby('Nama Pertemuan')['Jumlah Mahasiswa Lulus'].sum()

# Warna untuk setiap bar sesuai dengan ketentuan
colors = ['red', 'green', 'yellow', 'gray', 'blue', 'lightcoral']

# Nama pertemuan
labels = data_to_plot.index

# Plot diagram batang
plt.figure(figsize=(10, 10))
plt.bar(labels, data_to_plot, color=colors)
plt.title('Jumlah Mahasiswa Lulus Berdasarkan Pertemuan')
plt.xlabel('Nama Pertemuan')
plt.ylabel('Jumlah Mahasiswa Lulus')
plt.show()
```

# Soal No 6

Buatlah Layar Keluaran visualisasi Diagram Garis Berdasarkan Data File Excel "datamahasiswalulus"

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari file Excel
file_path = '../project-tugas/datamahasiswalulus.xlsx'  # Gantilah dengan path sesuai lokasi file Anda
df = pd.read_excel(file_path)

# Tentukan data yang akan diplot (misalnya, jumlah mahasiswa lulus per pertemuan)
data_to_plot = df.groupby('Nama Pertemuan')['Jumlah Mahasiswa Lulus'].sum()

# Nama pertemuan
labels = data_to_plot.index

# Plot diagram garis
plt.figure(figsize=(10, 10))
plt.plot(labels, data_to_plot, marker='o', linestyle='-', color='b', label='Jumlah Mahasiswa Lulus')

# Tambahkan judul dan label sumbu
plt.title('Jumlah Mahasiswa Lulus Berdasarkan Pertemuan')
plt.xlabel('Nama Pertemuan')
plt.ylabel('Jumlah Mahasiswa Lulus')

# Tampilkan legenda (opsional)
plt.legend()

# Tampilkan grid (opsional)
plt.grid(True)

# Tampilkan plot
plt.show()
```

# Soal No 8

Buatlah contoh kasus dari penggunaan scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Contoh data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 4, 6]

# Membuat scatter plot
plt.scatter(x, y)

# Menambahkan judul dan label sumbu
plt.title('Contoh Scatter Plot')
plt.xlabel('Variabel X')
plt.ylabel('Variabel Y')

# Menampilkan grafik
plt.show()
```

Buatlah contoh kasus dari penggunaan histogram

```python
import matplotlib.pyplot as plt

# Contoh data
data = [65, 30, 43, 49, 20, 35, 55, 60, 42, 48, 28, 38, 50]

# Membuat histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='red')

# Menambahkan judul dan label sumbu
plt.title('Contoh Histogram')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')

# Menampilkan grafik
plt.show()
```
