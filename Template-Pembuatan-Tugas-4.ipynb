Berikut ini adalah data tugas mahasiswa untuk setiap pertemuan yang lulus.

2. Pertemuan 2 : 30 Mahasiswa
3. Pertemuan 3 : 43 Mahasiswa

Buatlah layar keluaran menggunakan visualisasi PIE Chart, dengan ketentuan : 
1. Tampilkan Nama Pertemuan
2. Tampilkan Prosentase

```python
import matplotlib.pyplot as plt

# Data pertemuan dan jumlah mahasiswa yang lulus
pertemuan = ['Pertemuan 1', 'Pertemuan 2', 'Pertemuan 3', 'Pertemuan 4', 'Pertemuan 5']
jumlah_lulus = [65, 30, 43, 49, 20]

# Membuat PIE Chart
plt.figure(figsize=(6, 6))
plt.pie(jumlah_lulus, labels=pertemuan, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Menambahkan judul
plt.title('Persentase Jumlah Mahasiswa Lulus per Pertemuan')

# Menampilkan PIE Chart
plt.show()
```

# Soal No 5

Buatlah Layar Keluaran visualisasi Diagram Batang Berdasarkan Data File Excel "datamahasiswalulus"

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari file Excel
file_path = '../project-tugas/datamahasiswalulus.xlsx'  # Ganti path sesuai lokasi file
df = pd.read_excel(file_path)

# Tentukan data yang akan diplot
data_to_plot = df.groupby('Nama Pertemuan')['Jumlah Mahasiswa Lulus'].sum()

# Warna untuk setiap bar sesuai dengan ketentuan
colors = ['red', 'green', 'yellow', 'gray', 'blue', 'lightcoral']

# Nama pertemuan
labels = data_to_plot.index

# Plot diagram batang
plt.figure(figsize=(10, 10))
plt.bar(labels, data_to_plot, color=colors)
plt.title('Jumlah Mahasiswa Lulus Berdasarkan Pertemuan')
plt.xlabel('Nama Pertemuan')
plt.ylabel('Jumlah Mahasiswa Lulus')
plt.show()
```

# Soal No 6

Buatlah Layar Keluaran visualisasi Diagram Garis Berdasarkan Data File Excel "datamahasiswalulus"

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari file Excel
file_path = '../project-tugas/datamahasiswalulus.xlsx'  # Ganti path sesuai lokasi file
df = pd.read_excel(file_path)

# Tentukan data yang akan diplot 
data_to_plot = df.groupby('Nama Pertemuan')['Jumlah Mahasiswa Lulus'].sum()

# Nama pertemuan
labels = data_to_plot.index

# Plot diagram garis
plt.figure(figsize=(8, 8))
plt.plot(labels, data_to_plot, marker='o', linestyle='-', color='b', label='Jumlah Mahasiswa Lulus')

# Tambahkan judul dan label sumbu
plt.title('Jumlah Mahasiswa Lulus Berdasarkan Pertemuan')
plt.xlabel('Nama Pertemuan')
plt.ylabel('Jumlah Mahasiswa Lulus')

# Tampilkan legenda
plt.legend()

# Tampilkan grid
plt.grid(True)

# Tampilkan plot
plt.show()
```

# Soal No 8

Buatlah contoh kasus dari penggunaan scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
siswa = ['Siswa A', 'Siswa B', 'Siswa C', 'Siswa D', 'Siswa E']
waktu_belajar = [10, 8, 6, 7, 9]  # Jam
nilai_ujian = [90, 75, 60, 70, 85]

# Membuat scatter plot
plt.scatter(waktu_belajar, nilai_ujian, color='blue', alpha=1)

# Menambahkan label untuk setiap titik data 
for i, label in enumerate(siswa):
    plt.text(waktu_belajar[i], nilai_ujian[i], label)

# Menambahkan judul dan label sumbu
plt.title('Hubungan antara Waktu Belajar dan Nilai Ujian')
plt.xlabel('Waktu Belajar (jam)')
plt.ylabel('Nilai Ujian')

# Menampilkan garis tren sebagai ilustrasi
tren = np.polyfit(waktu_belajar, nilai_ujian, 1)
plt.plot(waktu_belajar, np.polyval(tren, waktu_belajar), color='red', linestyle='-', linewidth=1)

# Menampilkan grafik
plt.show()
```

Buatlah contoh kasus dari penggunaan histogram

```python
import matplotlib.pyplot as plt

# Data contoh: Skor ujian siswa
skor_ujian = [65, 75, 80, 90, 85, 70, 78, 94, 88, 69,
              73, 95, 83, 76, 87, 79, 98, 67, 91, 85]

# Membuat histogram
plt.hist(skor_ujian, bins=10, color='skyblue', edgecolor='black')

# Menambahkan judul dan label sumbu
plt.title('Distribusi Skor Ujian')
plt.xlabel('Skor Ujian')
plt.ylabel('Frekuensi')

# Menampilkan grid (opsional)
plt.grid(axis='y', linestyle='--', alpha=1)

# Menampilkan grafik
plt.show()
```
